```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("fetal_health.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
corrmat=data.corr() 
top_corr_features=corrmat.index 
plt.figure(figsize=(20,20)) #plot heat map 
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
data.head()
```

```python
x=data.drop(['fetal_health'],axis=1)
y=data['fetal_health']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
dc=DecisionTreeClassifier()
dc.fit(X_train,y_train)
```

```python
X_train.shape
```

```python
dc_pred=dc.predict(X_test)
```

```python
X_test.shape
```

```python
print(accuracy_score(dc_pred,y_test))
```

```python
print(classification_report(dc_pred,y_test))
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
model=SVC(kernel='rbf', C=30, gamma='auto')
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
param_grid = {'kernel': ('linear', 'rbf'),
              'C': [1, 10, 20]}
base_estimator = SVC(gamma='auto')
```

```python
gs=GridSearchCV(base_estimator, param_grid,cv=5)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_params_
```

```python
gs.best_estimator_
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
```

```python
gbc = GradientBoostingClassifier()
parameters = {
    "n_estimators":[5,50,250,500],
    "max_depth":[1,3,5,7,9],
    "learning_rate":[0.01,0.1,1,10,100]
}
```

```python
cv = GridSearchCV(gbc,parameters,cv=5).fit(X_train,y_train)
```

```python
cv.best_score_
```

```python
cv.best_params_
```

```python
cv_pred=cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(cv_pred,y_test))
```

```python
print(classification_report(cv_pred,y_test))
```
